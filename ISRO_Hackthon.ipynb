```python
# STEP 1: Install Ultralytics
!pip install ultralytics --quiet

# STEP 2: Import dependencies
import cv2
import numpy as np
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import display
from google.colab import files

# STEP 3: Load your trained model (update this path if needed)
model_path = "/content/yolo_results/moon_crater_boulder_detector/weights/best.pt"
model = YOLO(model_path)

# STEP 4: Upload an image
uploaded = files.upload()
image_path = next(iter(uploaded))

# STEP 5: Run inference
results = model(image_path, conf=0.2)  # lowered confidence threshold
result = results[0]

# STEP 6: Load image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]

# STEP 7: Extract detections
boulders = []
craters = []

for box in result.boxes.data.cpu().numpy():
    x1, y1, x2, y2, conf, cls_id = box
    cls_id = int(cls_id)
    cx, cy = int((x1 + x2) / 2), int((y1 + y2) / 2)

    if result.names[cls_id] == "boulder":
        boulders.append((cx, cy))
    elif result.names[cls_id] == "crater":
        craters.append((cx, cy, int((x2 - x1) / 2)))  # include radius estimate

# STEP 8: Detect landslide zones
landslide_zones = []

# Rule 1: Boulders close to each other (<30 pixels)
for i in range(len(boulders)):
    for j in range(i + 1, len(boulders)):
        if np.linalg.norm(np.array(boulders[i]) - np.array(boulders[j])) < 30:
            mid = tuple(np.mean([boulders[i], boulders[j]], axis=0).astype(int))
            landslide_zones.append(mid)

# Rule 2: Boulders near crater edge (within ~5-15 pixels)
for boulder in boulders:
    for cx, cy, r in craters:
        dist = np.linalg.norm(np.array([cx, cy]) - np.array(boulder))
        if r - 15 < dist < r + 15:
            landslide_zones.append(boulder)

# Rule 3: Overlapping craters (center distance < avg radius)
for i in range(len(craters)):
    for j in range(i + 1, len(craters)):
        (x1, y1, r1), (x2, y2, r2) = craters[i], craters[j]
        dist = np.linalg.norm(np.array([x1, y1]) - np.array([x2, y2]))
        if dist < (r1 + r2) * 0.6:  # high overlap
            mid = tuple(np.mean([[x1, y1], [x2, y2]], axis=0).astype(int))
            landslide_zones.append(mid)

# STEP 9: Draw semi-transparent bounding boxes
for box in result.boxes.data:
    x1, y1, x2, y2, conf, cls_id = box.tolist()
    x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
    cls_id = int(cls_id)
    label = f"{result.names[cls_id]} {int(conf * 100)}%"

    color = (0, 255, 0) if result.names[cls_id] == "boulder" else (0, 0, 255)
    overlay = img_rgb.copy()
    cv2.rectangle(overlay, (x1, y1), (x2, y2), color, -1)
    alpha = 0.2
    img_rgb = cv2.addWeighted(overlay, alpha, img_rgb, 1 - alpha, 0)

    (text_w, text_h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
    cv2.rectangle(img_rgb, (x1, y1 - text_h - 4), (x1 + text_w, y1), color, -1)
    cv2.putText(img_rgb, label, (x1, y1 - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

# STEP 10: Plot landslide zones with red dots
for pt in landslide_zones:
    cv2.circle(img_rgb, pt, 4, (255, 0, 0), -1)  # red dot

# STEP 11: Save and show result
output_path = "/content/landslide_output.png"
cv2.imwrite(output_path, cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
display(Image.open(output_path))
```

```python
# STEP 1: Install Ultralytics
!pip install ultralytics --quiet

# STEP 2: Import dependencies
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from PIL import Image
from google.colab import files

# STEP 3: Load your trained model
model_path = "/content/yolo_results/moon_crater_boulder_detector/weights/best.pt"
model = YOLO(model_path)

# STEP 4: Upload image
uploaded = files.upload()
image_path = next(iter(uploaded))

# STEP 5: Run inference
results = model(image_path, conf=0.1)
result = results[0]

# STEP 6: Load and prepare image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# STEP 7: Collect detections
boulders = []
craters = []

for box in result.boxes.data.cpu().numpy():
    x1, y1, x2, y2, conf, cls_id = box
    cls_id = int(cls_id)
    cx, cy = int((x1 + x2) / 2), int((y1 + y2) / 2)
    radius = int((x2 - x1) / 2)

    # Assign detection by class
    if result.names[cls_id] == "boulder":
        boulders.append((cx, cy))
        color = (0, 255, 0)  # Green (BGR)
    elif result.names[cls_id] == "crater":
        craters.append((cx, cy, radius))
        color = (0, 255, 255)  # Cyan (BGR)

    x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
    cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color, 2)

# STEP 8: Detect possible landslides
landslide_zones = []

# Rule 1: Boulders close to each other
for i in range(len(boulders)):
    for j in range(i + 1, len(boulders)):
        if np.linalg.norm(np.array(boulders[i]) - np.array(boulders[j])) < 30:
            mid = tuple(np.mean([boulders[i], boulders[j]], axis=0).astype(int))
            landslide_zones.append(mid)

# Rule 2: Boulders near crater edge
for bx, by in boulders:
    for cx, cy, r in craters:
        dist = np.linalg.norm(np.array([cx, cy]) - np.array([bx, by]))
        if r - 15 < dist < r + 15:
            landslide_zones.append((bx, by))

# Rule 3: Overlapping craters
for i in range(len(craters)):
    for j in range(i + 1, len(craters)):
        (x1, y1, r1), (x2, y2, r2) = craters[i], craters[j]
        dist = np.linalg.norm(np.array([x1, y1]) - np.array([x2, y2]))
        if dist < (r1 + r2) * 0.6:
            mid = tuple(np.mean([[x1, y1], [x2, y2]], axis=0).astype(int))
            landslide_zones.append(mid)

# STEP 9: Draw red dots for sliding zones
for pt in landslide_zones:
    cv2.circle(img_rgb, pt, 4, (255, 0, 0), -1)  # Red

# STEP 10: Display image with legend
legend_elements = [
    Patch(facecolor='cyan', edgecolor='black', label='Crater'),
    Patch(facecolor='green', edgecolor='black', label='Boulder'),
    Patch(facecolor='red', edgecolor='black', label='Sliding Zone')
]

plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.axis('off')
plt.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=3, frameon=False)
plt.tight_layout()
plt.show()
```
